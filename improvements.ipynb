```python
from collections import OrderedDict
```

```python

with open("mr_cmupan-ud-train.conllu", "r", encoding="utf-8") as data_file:
    sentences = []
    i=0
    for tokenlist in parse_incr(data_file):

        tokenlist.metadata["sent_id"] = str(i)  # sent_id is usually a string
        sentences.append(tokenlist)
        tokenlist.metadata=OrderedDict([
            ("sent_id",str(i)),
            ("text",tokenlist.metadata.get("text",""))
        ])
        i+=1
with open("updated_train_set.conllu", "w", encoding="utf-8") as out_file:
    for tokenlist in sentences:
        out_file.write(tokenlist.serialize())
```

### NST tag issue

```python
from conllu import parse_incr,parse,parse_tree

with open ("/Users/pranavk/Desktop/HiWi/Marathi Treebanking/Code/UD_Marathi-CMUPAN/final_dev.conllu" ,"r") as f:
    data=f.read()

parsed_sentences=parse(data)
print(parsed_sentences[0])
```

```python
folders=["final_dev","final_train","final_test"]

NST_Words=[]
for split in folders:
    with open (f"/Users/pranavk/Desktop/HiWi/Marathi Treebanking/Code/UD_Marathi-CMUPAN/{split}.conllu" ,"r") as f:
        data=f.read()
        
    parsed_sentences=parse(data)
    for sent in parsed_sentences:
        for tok in sent:
            if tok["upos"]=="NST":
                NST_Words.append(tok["form"])
```

```python
len(NST_Words)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df=pd.DataFrame(pd.DataFrame(NST_Words).value_counts())
df=df.reset_index()
df
```

```python
df.columns=["word","count"]
df
```

```python
from googletrans import Translator
import asyncio
import tqdm
translated=[]

for words in tqdm.tqdm(df["word"].values):

    async with Translator() as translator:
            result = await translator.translate(words,src="mr")
            translated.append(result.text)
```

```python
df["Translation"]=translated
df
```

```python
df.to_csv("NST_tag_words.csv")
```

### Replacing NST TAG

```python
from conllu import parse_incr,parse,parse_tree
import nltk
```

```python
#https://github.com/verachell/English-word-lists-parts-of-speech-approximate/blob/main/other-categories/mostly-prepositions.txt


with open("/Users/pranavk/Desktop/HiWi/Marathi Treebanking/Code/UD_Marathi-CMUPAN/files/mostly-prepositions.txt") as f:
    pos=f.read().split("\n")
```

```python
import pandas as pd
df=pd.read_csv("/Users/pranavk/Desktop/HiWi/Marathi Treebanking/Code/UD_Marathi-CMUPAN/NST_tag_words.csv")
df["pos"]=""
df.head()
```

```python
for id,rows in df.iterrows():
    if rows["Translation"].lower() in pos:
        df.at[id,"pos"]="ADP"
```

```python
for id,row in df[df["pos"]!="ADP"].iterrows():
    tag=nltk.pos_tag([row["Translation"]])
    if "NN" in str(tag):
        df.at[id,"pos"]="NOUN"
```

```python
for id,row in df[df["pos"]==""].iterrows():
    df.at[id,"pos"]="ADV"
```
